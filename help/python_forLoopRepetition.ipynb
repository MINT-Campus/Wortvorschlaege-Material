# for-Schleifen

### Motivation
Eine **for-Schleife** hilft uns eine Anweisung mehrmals auszuführen ohne viel Schreibaufwand! Angenommen wir wollen die Gewichtsangaben in einem Rezept halbieren, weil wir nur die Hälfte der Menge zubereiten wollen. Das können wir bei einer kleinen Anzahl an Zutaten noch gut per Hand bestimmen:  
(*Klicke in das folgende Code-Feld und drücke auf "Run".*)

````python
weights= [200,150,175];
# Das Gewicht der Zutaten halbieren:
half_weights = [200/2, 150/2, 175/2];
print(f"Das halbierte Gewicht der Zutaten ist {half_weights}")
````

Wollen wir jedoch von mehr als nur drei Zutaten das Gewicht halbieren, oder allgemein eine Rechenanweisung für mehrere Listenelemente durchführen, so wird dies per Hand schon ziemlich aufwendig.   
Zum Glück gibt es sogenannte **for-Schleifen**. Diese dienen dazu, einen bestimmten Abschnitt des Codes z. B. eine bestimmte Rechenoperation - in unserem Beispiel das Halbieren jedes Eintrages der Zutatenliste - mehrmals hintereinander auszuführen.

### Formal:
Werfen wir zuerst einen Blick auf die formale Syntax:

```python
for Variable in Liste:
    Anweisung(en)
```

In der ersten Zeile legt man den Namen der Schleifenvariable und die Liste, die die Schleifenvariable durchlaufen soll, fest. Die Schleifenvariable (hier `Variable` genannt) wird zu Anfang auf das erste Element der Liste gesetzt und dann in jedem Durchgang auf das Element rechts nebendran, bis die Liste komplett durchlaufen ist. Es gibt auch die Möglichkeit, anstatt der Liste einen Start und Endwert anzugeben. Die Schleifenvariable startet dann beim Startwert und wird in jedem Durchgang um eine angegebene Schrittweite erhöht bis der Endwert erreicht ist. Hierauf gehen wir allerdings nicht weiter ein.

### Am Beispiel:
Mithilfe einer for-Schleife sieht die Halbierung der Einträge dann so aus:  
(*Klicke in das folgende Code-Feld und drücke auf "Run".*)

````python
# Das Gewicht der Zutaten halbieren:

weights = [200, 150, 175, 80, 75, 120, 250, 300];

for i in weights:
    i = i/2;
    print(f"Das halbierte Gewicht ist {i}")
````

In Worten kann die for-Schleife so beschrieben werden: 
* Setze i auf den Wert des nullten Eintrages der Liste `weights`: $i = 200$.
* Halbiere i. Die Variable `i` hat nun den Wert $i = 200 : 2 = 100$.
* Der Wert für i in diesem Schritt wird ausgegeben.
* Setze i nun auf den Wert des ersten Eintrages der Liste `weights`: $i = 150$.
* Halbiere i. Die Variable `i` hat nun den Wert $i = 150 : 2 = 75.5$.
* Der Wert für i in diesem Schritt wird ausgegeben.
* Setze i nun auf den Wert des zweiten Eintrages der Liste `weights`: $i = 175$.
* Halbiere i. Die Variable `i` hat nun den Wert $i = 175 : 2 = 87,5$.
* Der Wert für i in diesem Schritt wird ausgegeben.
*  $\dots$
* Dieser Vorgang wird solange wiederholt, bis i jeden Eintrag der Liste `weights` durchlaufen hat.
