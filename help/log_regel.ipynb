# Logarithmusgesetze

Für das Rechnen mit Logarithmen gelten folgende Regeln:

### Produktregel
Der Logarithmus eines Produktes ist die Summe der Logarithmen der beiden Faktoren:</br>

<center>$log(a\cdot b)=log(a)+log(b)$</center>

### Quotientenregel
Der Logarithmus eines Bruchs ist der Logarithmus des Zählers abzüglich des Logarithmus des Nenners:</br>

<center>$log(\frac{a}{b})=log(a)-log(b)$</center>

### Potenzregel

Der Logarithmus einer Potenz ist das Produkt aus Exponenten und Logarithmus der Basis der Potenz:</br>

<center>$log(a^n)=n \cdot log(a)$</center>
