Um zu überprüfen, ob die letzten beiden Wörter/ das letzte Wort im Datensatz enthalten sind, kannst du den Befehl <SPAN STYLE="font-family:'Monaco'; font-size: 9.5pt; background-color: #EEEEEE"> <span style="color:blue">w</span> in find(<span style="color:blue">T</span>)</SPAN> verwenden. Setze für <SPAN STYLE="font-family:'Monaco'; font-size: 9.5pt; background-color: #EEEEEE"><span style="color:blue" > T</span></SPAN> die Tabelle zum jeweiligen Modell ein (entweder `T_Tri`oder `T_Bi`) und für <SPAN STYLE="font-family:'Monaco'; font-size: 9.5pt; background-color: #EEEEEE"> <span style="color:blue">w</span></SPAN> die letzten beiden Wörter/ das letzte Wort.
