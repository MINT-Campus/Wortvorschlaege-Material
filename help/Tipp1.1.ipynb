# Tipp

Die for-Schleife sorgt dafür, dass Wort1 und Wort2 alle Wörter des Datensatzes durchlaufen. Im ersten Durchlauf nimmt `Wort1` den Wert "ich" an und `Wort2`den Wert "ich". Im zweiten Durchlauf wird `Wort2` auf "liebe" gesetzt. Sodass alle mögliche Kombinationen für zwei Wörter des Datensatz durchlaufen werden. In jedem Schritt können wir somit einen Eintrag der Tabelle 1 befüllen.
