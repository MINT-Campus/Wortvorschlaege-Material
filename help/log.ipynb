# Logarithmusfunktion

Die (natürliche) Logarithmusfunktion ist die Umkehrfunktion der Exponentialfunktion.

### Eigenschaften der Logarithmusfunktion
Sie hat im Punkt P(1|0) einen Schnittpunkt mit der x-Achse.

Sie verläuft ausschließlich im ersten und vierten Quadranten.

Die y-Achse, also die Grade mit der Gleichung x=0 ist die einzige Asymptote dieser Funktion. Die Funktion nähert sich stets der y-Achse an, wenn die x-Werte gegen Null gehen, schneidet sie aber nicht.

Die Logarithmusfunktion ist streng monoton steigend.

Den Definitionsbereich für diese Funktionen bilden alle x-Werte, die größer als Null sind: $D=\{\mathbb{R}, x>0\}$. Der Wertebereich sind alle reellen y-Werte: $W=\mathbb{R}$.

### Die Logarithmusfunktion in Python

Auf die Logarithmusfunktion kann in Python mit `np.log(x)` zugegriffen werden.

### Anwendung und Ausblick
Wendet man die Logarithmusfunktion auf eine Funktion an, so ist sie zugleich eine Rechenoperation, mit der man den (gesuchten) Exponenten einer bekannten Zahl herausfinden kann.</br>
Der Logarithmus wird außerdem in der Numerik häufig verwendet. Denn der Logarithmus angewendet auf eine Funktionen wandelt Produkte in Summen um und erleichtert so das weitere Rechnen mit dieser Funktion.</br>
