# Tipp

Um den Übergangsgraphen zu bilden ist es hilfreich, alle Bi-Gramme im Text <span style="color:teal">"Ich mag Physik. Ich mag Sport. Ich liebe Mathe. Jeder mag Ferien."</span > einzukreisen und die Bi-Gramme in die Wortvorgeschichte (das erste Wort des Bi-Grammes) und das Folgewort (das letzte Wort des Bi-Grammes) zu unterteilen. Ordne anschließend die Wortvorgeschichte und das Folgewort als Knoten im Graphen in der richtigen Reihenfolge an. Für die ersten drei Sätz sieht der Übergangsgraph beispielsweiße wie folgt aus:
