# if-Bedingung

### Motivation
Angenommen wir wollen eine Berechnung oder Anweisung nur dann duchführen, wenn eine bestimmte Bedingung erfüllt ist. Dann könnten wir das mit einer sogenannten if-Bedingung tun.

### Formal:
Werfen wir zuerst einen Blick auf die formale Syntax:

```python
if Bedingung I:
    Anweisung(en) I
    
elif Bedingung II:
    Anweisung(en) II

elif Bedingung III:
    Anweisung(en) III
    
 ... usw...
    
elif Bedingung X:
    Anweisung(en) X
```
In Worten kann man die if-Bedingung so beschreiben:
* In der ersten Zeile wird die I Bedingung eingegeben.
* Ist Bedingung I  erfüllt? 
    * JA:   Anweisung I wird ausgeführt. STOP
    * NEIN: Anweisung I wird nicht ausgeführt. Weiter gehts zu Bedingung II.
* Ist Bedingung II  erfüllt?  
    * JA:   Anweisung II wird ausgeführt. STOP
    * NEIN: Anweisung II wird nicht ausgeführt. Weiter gehts zu Bedingung III.
* wenn bisher keine Bedingung erfüllt wurde, werden nach und nach alle weiteren Bedingungen durchgegangen. 

**Hinweis:** Schreibt man `else` anstelle des letzten `elif`, so führt Python die zugehörige Anweisung aus, wenn alle bisherigen Bedingungen nicht erfüllt waren.

### Beispiel:
* Der Wert der Variable `a` soll quadriert werden,  **wenn** `a` negativ ist. 
* Der Wert der Variable `a` soll um 12 erhöht werden,  **wenn** `a` Null ist. 
* Der Wert der Variable `a` soll halbiert werden,  **wenn** `a` positiv ist.

(*Klicke in das folgende Code-Feld und drücke auf "Run".*)

````python
# Speichern eines Wertes für die Variable a
a = -2; # Ändere den Wert von a und beobachte das Ergebnis.

if a < 0:
    a = a^2;
    
elif a == 0:
    a = a + 12;
    
else: # Hier hätten wir auch elif a > 0 schreiben können
    a = a / 2;
    
print(f"Der Wert von a ist {a}")
````
