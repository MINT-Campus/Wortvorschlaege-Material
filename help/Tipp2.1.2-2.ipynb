# Tipp2 | Übergangsgraph des Tri-Gramm-Modells

 Die grundlegende Struktur des Übergangsgraphen sollte nach dem Hinzufügen aller Wortvorgeschichten und Folgewörter wie folgt aussehen:
<div align="center">
<img src="../figs/Tipp1-2.JPG"style="width: 200px;"/> 
</div>
Um nun die relativen Häufigkeiten $h_\text{Wort 0, Wort 1}(\text{Wort 2})$ bestimmen zu können, benötigen wir die Auftretenshäufigkeit der Wortsequenz $\text{N(Wort 0, Wort 1, Wort 2)}$, sowie die Summe der Auftretenshäufigkeit der Wortsequenzen $\text{N(Wort 0, Wort 1, beliebiges Wort)}$. Die relative Häufigkeit ergibt sich zu
&nbsp;

&nbsp;

<div align="center">
$h_\text{Wort 0, Wort 1}(\text{Wort 2})=\frac{N(\text{Wort 0, Wort 1, Wort 2})}{N(\text{Wort 0, Wort 1, beliebiges Wort 1})+N(\text{Wort 0, Wort 1, beliebiges Wort 2})+...}$.
</div>

&nbsp;

**Beispiel:** <br />
Die relative Häufigkeit $h_\text{ich, mag}(\text{Physik})$ berechnet sich zu <br />
<div align="center">
$h_\text{ich, mag}(\text{Physik})=\frac{N(\text{Ich, mag, Physik})}{N(\text{Ich, mag, Physik})+N(\text{Ich, mag, Sport})}=\frac{1}{1+1}=\frac{1}{2}$.
</div>
&nbsp;

Dabei wurden alle Auftretenshäufigkeiten, die null sind, direkt weggelassen.
