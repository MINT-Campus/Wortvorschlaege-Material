## Teil 1) Bewertungskriterien für Wortvorhersagemodelle
____
<i class="fa fa-pencil-square-o" style="font-size:38px"></i>  &     <i class="fa fa-comments-o" style="font-size:36px"></i> </br>

_Welche Eigentschaft muss ein Wortvorhersagemodell erfüllen, damit es ein gutes Wortvorhersagemodell ist? Überlege dir, welche Testungen wir durchführen können, um die Qualität eines Vorhersagemodells festzustellen._

<details>
  <summary><b>Lösung</b> (Erst die Frage beantworten, dann hier klicken!)</summary>
&nbsp;

Ein Wortvorhersagemodell ist umso besser, je häufiger es das gewünschte Wort vorschlägt und die Nutzenden somit beim Tippen unterstützt.</br>
Wir können auf verschiedene Weisen feststellen, wie gut ein Sprachmodell diese Eigenschaft erfüllt:
&nbsp;

* **Zeitersparnis**: Wir können die Zeit messen, die beim Tippen eines Textes mit Verwendung eines Vorhersagemodell im Vergleich zum Tippen ohne Wortvorschläge eingespart wird. Die Zeitersparnis kann allerdings nur mit Hilfe von Versuchspersonen festgestellt werden und ist für uns daher weniger geeignet.
* **Einsparung der Tastenanschläge**: Wir können zählen, wie viele Tastenanschläge durch die Nutzung der Wortvorschläge eingespart werden.
* **Trefferrate**: Wir können zählen, wie häufig eines der vorgeschlagenen Wörter tatsächlich ausgewählt wird. Wir teilen dies durch die Anzahl der insgesamt geschriebenen Wörter.
* **Kreuzentropie**: Wir können die (durchschnittliche logarithmierte) Wahrscheinlichkeit, mit der ein geschriebener Text aus unserem Sprachmodell entstehen würde, berechnen. Je wahrscheinlicher der Text, generiert aus unserem Sprachmodell entstehen würde, desto besser ist das Modell.
 
</details>
&nbsp;


____

&nbsp;


Wie können wir diese Testungen nun durchführen? Werden dafür tatsächlich immer Versuchspersonen gebraucht um das Modell zu testen oder gibt es einen einfacheren Weg?</br>
Bei der Trefferrate können beispielsweiße bereits geschriebene Texte auf die Übereinstimmung mit den gegebenen Wortvorschlägen auf der Grundlage der vorherigen Wörter im Text überprüft werden. Der Text wird also komplett durchlaufen und nach jedem Wort wird überprüft, ob das Folgewort mit einem der drei Wortvorschläge übereinstimmt oder nicht.</br>Diese Texte, mit denen hier gearbeitet wird, werden zumeist aus der Datengrundlage, aus welcher auch die Trainingstexte stammen, entnommen. Ein Teil der Texte wird somit zur Bestimmung der Übergangswahrscheinlichkeiten genutzt und die übrigen Sätze werden zur Bewertung des Modells genutzt. Diese Methode wird als "Hold-Out"-Methode bezeichnet, da wir ein paar zufällig ausgewählte Sätze aus dem Trainingsdatensatz für die spätere Bewertung zurückhalten. Es werden daher keine Versuchsperson benötigt, um das Modell zu bewerten, sondern wir können sogenannte <span style="color: firebrick;">**Testdatensätze**<span style="color: black;"> verwenden. Diese umfassen wie die Trainingsdatensätze auch mehrere tausend Wörter. Das ist wichtig damit die Bewertung auch die tatsächliche Qualität des Vorhersagemodells widerspiegelt. Bei wenigen Wörtern könnte es auch nur Zufall sein, dass das Modell gut oder schlecht abgeschnitten hat. Der Testdatensatz muss daher eine gewisse Größe haben, damit man sich auf das Ergebnis der Bewertung verlassen kann.

In der Praxis wird vor allem die Kreuzentropie genutzt. Sie hat die folgenden Vorteile gegenüber der anderen Bewertungskriterien:
* Die Berechnung der Zeitersparnis, Trefferrate oder Einsparung der Tastenanschläge erfordert die Generierung der Wortvorschläge. Bei unseren Modellen benötigt die Generierung des Vorschlages an sich wesentlich mehr Zeit als die bloße Bildung der Übergangswahrscheinlichkeiten. Bei einem Bewertungstext von mehreren tausend Wörtern dauert es daher sehr lange, bis für jedes Wort der Vorschlag generiert wird und im Anschluss beispielsweise die Trefferrate berechnet wird. Für die Berechnung der Kreuzentropie werden nur die Übergangswahrscheinlichkeiten benötigt, wodurch die Berechnung sehr schnell ist.
* Im Allgemeinen möchte man Sprachmodelle mit anderen Sprachmodellen vergleichen, die nicht nur zur Wortvorhersage, sondern auch für andere Zwecke eingesetzt werden. Zum Sprachmodell selbst gehören nur die Übergangstabellen ohne die Generierung eines Vorschlages. In der Spracherkennung werden diese Übergangswahrscheinlichkeiten beispielsweiße in den Erkennungsprozess eines gesprochenen Textes mit einbezogen. Der Wortvorschlag kann für diesen Zweck nicht verwendet werden. Die obigen Bewertungskriterien sind jedoch nur für Wortvorhersagemodelle sinnvoll. Einzig die Kreuzentropie spiegelt die Bewertung des Sprachmodells und nicht, wie die anderen Bewertungskriterien, die Bewertung des Wortvorhersagegemodell unter Verwendung des Sprachmodells wider.

## Teil 2) Kreuzentropie als Bewertungsmaß
Wir wollen nun unser letztes Modell, das kombinierte N-Gramm-Modelle mit gewichtetem Mittelwert, bewerten. Dazu schauen wir uns die **Kreuzentropie** einmal genauer an. Diese ist über die Formel </br>
<center>$K=\color{Orange}{-\frac{1}{N}}\cdot \color{SeaGreen}{\log} (\color{Maroon}{\mathbb{P}(\text{Wort 1,Wort 2,Wort 3,Wort 4,...})})$</center>
definiert.
&nbsp;

* $\color{Maroon}{\mathbb{P}(\text{Wort 1,Wort 2,Wort 3,Wort 4,...})}$ steht hierbei für die Wahrscheinlichkeit mit der aus unserem Modell ein Text bestehend aus den Wörtern $\text{Wort 1,Wort 2,Wort 3,Wort 4,...}$ entsteht. Sie wird auch als **Modellwahrscheinlichkeit** bezeichnet und spiegelt die Anpassung des Modells an den Testdatensatz wider. </br>
* $\color{SeaGreen}{\log}$ bezeichnet die Logarithmusfunktion. </br>

**<i class="fas fa-info-circle" style="font-size:32px"></i> Falls du die Logarithmusfunktion, noch nicht kennst, kannst du dir [hier](../help/log.ipynb) die Information zur Logarithmusfunktion ansehen.** </br>
* Multipliziert wird der Logarithmus der Modellahrscheinlichkeit mit dem Faktor $\color{orange}{-\frac{1}{N}}$. Hierbei steht $N$ für die Anzahl der Wörter im Testdatensatz. Somit skaliert der Faktor das Bewertungsmaß $K$ mit der Größe des Testdatensatzes.</br>

Die Kreuzentropie $K$ ist also umso niedriger, und die Anpassung des Modells an den Testdatensatz umso besser, je höher die
Modellwahrscheinlichkeit des Testdatensatzes $\text{Wort 1,Wort 2,Wort 3,Wort 4,...}$ ist. Mit anderen Worten: dasjenige Modell ist am besten, welches dem Testdatensatz die höchste Wahrscheinlichkeit zuweist.</br>
____
**Merke:** Eine kleine Kreuzentropie bedeutet, dass das Sprachmodell zu unserem Testdatensatz passt und somit für diese Art von Texten die Übergangswahrscheinlichkeiten gut schätzt.
