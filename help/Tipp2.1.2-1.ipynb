# Tipp1 | Übergangsgraph des Tri-Gramm-Modells

Da eine Wortvorgeschichte nun aus zwei Wörtern besteht, ist es notwendig, zwei aufeinanderfolgende Wörter als einen Knoten im Graphen zu behandeln.\
Zum Beispiel werden die Worte "Ich" und "mag" als eine gemeinsame Wortvorgeschichte betrachtet:
<div align="center">
<img src="../figs/Tipp1.JPG"style="width: 250px;"/> 
</div>

Bilde den Übergangsgraphen indem du wie im Beispiel der Wortvorgeschichte "Ich mag" immer zwei Wörter als einen Knoten im Übergangsgraphen auffasst. Kreise dazu alle Tri-Gramme im Text ein und unterteile die Tri-Gramme in die Wortvorgeschichte (die ersten beiden Wörter des Tri-Grammes) und das Folgewort (das letzte Wort des Tri-Grammes). Ordne anschließend die Wortvorgeschichte und das Folgewort als Knoten im Graphen in der richtigen Reihenfolge an. Für den ersten Satz sieht der Übergangsgraph beispielsweise wie folgt aus:
<div align="center">
<img src="../figs/Tipp1_2.JPG"style="width: 280px;"/> 
</div>
Füge nach diesem Schema alle Tri-Gramme zum Übergangsgraphen hinzu.
&nbsp;


Anmerkung: Es ist ebenfalls korrekt, das Bi-Gramm "mag Physik" als rechten Knoten zu wählen anstatt nur "Physik". Dies ist vor allem dann wichtig, wenn der Satz nach diesen Wörtern noch nicht zu Ende ist und das Bi-Gramm "mag Physik" selbst wieder Vorgeschichte für ein weiteres Wort ist. Da hier der Satz nach dem Wort "Physik" endet, hat das Hinzufügen des Wortes "mag" zum rechten Knoten keinen weiteren Nutzen und kann somit weggelassen werden.
