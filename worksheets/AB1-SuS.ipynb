## 2) allgemeine Formel bestimmen

Bei unserem kleinen Trainingsdatensatz ist es noch einfach, die relativen Häufigkeiten per Hand zu ermitteln. Für größere Datensätze ist das jedoch viel zu aufwändig. Entwickle deshalb eine **allgemeine Formel für die relative Häufigkeit**, sodass die Berechnung automatisiert werden kann. Bearbeite dazu folgende Aufgaben:

<div id="exercise" class="alert alert-info">
    <h3><i class="fa fa-pencil-square-o" style="font-size:30px"></i> <br><br> a) relative Häufigkeit für ein Beispiel </h3>
    Stelle eine Formel für die relative Häufigkeit auf, mit der das Wort "mag" nach dem Wort "ich" auftritt. Wir wollen diese relative Häufigkeit mit $h_\text{ich}(\text{mag})$ bezeichnen.
    <hr>
     <i class="fas fa-wrench" style="font-size: 20px"></i> &nbsp;  <strong>Schreibweise:</strong> 
    Im Workshop wird ab hier mit den klein geschriebenen Wörtern gearbeitet.
    <hr>
    <details><summary><i class="fas fa-life-ring" style="font-size:20px"></i> &nbsp;Tipp 1 </a> <span style="margin-right: 20px;"></span></summary>
    <br>
Die relative Häufigkeit $h_\text{ich}(\text{mag})$ kann auch als Häufigkeit aufgefasst werden, mit der die Wortsequenz "ich mag" unter allen Wortsequenzen mit "ich" und einem beliebigen Wort auftritt.
    <br>
    <br>
</details><details>
  <summary><i class="fas fa-life-ring" style="font-size:20px"></i> &nbsp;Tipp 2 </a> <span style="margin-right: 20px;"></span></summary>
    <br>
Bestimme zunächst die absolute Häufigkeit der Wortsequenz "ich mag".
    <br>
    <br>
</details>
    <hr>
    <details> 
  <summary><i class="fas fa-lightbulb" style="font-size:20px"></i> &nbsp;<b>Lösung</b> (Erst die Frage beantworten, dann hier klicken!)</summary>
<br>
Wir betrachten alle Zweier-Wortsequenzen im Datensatz mit erstem Wort "ich". Um die relative Häufigkeit $h_\text{ich}(\text{mag})$ bestimmen zu können, zählen wir, wie häufig die Wortsequenz $\text{(ich, mag)}$ auftritt (absolute Häufigkeit). Wir nennen diese Anzahl $\text{N(ich, mag)}$. Für den obigen Datensatz gilt $\text{N(ich, mag)}=2$.<br />
Außerdem zählen wir, wie häufig das Wort "ich" gefolgt von einem beliebigen Wort auftritt. Wir nennen diese Anzahl $\text{N(ich, beliebiges Wort)}$. Es gelten <br />
<div align="center">
$\text{N(ich, liebe)}=1$,  <br />
$\text{N(ich, ich)}=\text{N(ich, Mathe)}=\text{N(ich, Physik)}=\text{N(ich, Sport)}=0$.
</div> <br />
<br>
Die relative Häufigkeit ergibt sich als Quotient aus $\text{N(ich, mag)}$ und $\text{N(ich, beliebiges Wort)}$ zu <br><br>
<div align="center">
$h_\text{ich}(\text{mag})=\frac{N(\text{ich},\text{ mag})}{N(\text{ich},\text{ich})+N(\text{ich},\text{mag})+N(\text{ich},\text{liebe})+...}=\frac{2}{0+2+1+0+0+0}=\frac{2}{3}$.
</div>
&nbsp;<br>

Die bedingte Wahrscheinlichkeit ist die Wahrscheinlichkeit des Eintretens eines Ereignisses A unter der Bedingung, dass das Eintreten eines anderen Ereignisses B bereits bekannt ist, und man schreibt hierfür \mathbb{P}_\text{B}(\text{A}). In unserem Fall steht das Ereignis A für das Auftreten des Wortes "mag", und das Ereignis B für das Auftreten des Wortes "Ich", gefolgt von einem beliebigen Wort.-->

</div>

## 3) Die besten Wörter vorschlagen

Wir wollen die relativen Häufigkeiten nun nutzen, um Vorschläge für das nächste Wort zu geben. Nach Tippen des Wortes "ich" wollen wir dasjenige Wort vorschlagen, das am häufigsten beim Chatten nach dem Wort "ich" geschrieben wird. Mit dem Wissen um die relativen Häufigkeiten im Trainingsdatensatz würden wir das Wort "mag" nach dem Wort "ich" mit der höchsten relativen Häufigkeit von $\frac{2}{3}$ vorschlagen.

<div id="information" class="alert alert-success">
  <h3><i class="fas fa-info" style="font-size:30px"></i> Modellebene </h3>
    <ol>   
In unserem Modell fassen wir das Auswählen eines Wortes nach einem bereits geschriebenen Wort als Zufallsexperiment auf. Die Wahl des nächsten Wortes erfolgt entsprechend der relativen Häufigkeit $h_\text{Wort 1}(\text{Wort 2})$, mit der die Wortsequenz insgesamt in allen Texten, die beim Chatten verfasst wurden, auftritt. Wir nennen diese relative Häufigkeit <b> Übergangswahrscheinlichkeit</b>. Diese Übergangswahrscheinlichkeiten kennen wir allerdings nicht. Wir können nur diejenigen relativen Häufigkeiten beobachten, mit der die Wortsequenzen im Trainingsdatensatz auftreten. Wir hoffen, dass die relativen Häufigkeiten, die wir im Trainingsdatensatz beobachten, sich nicht zu stark von den Übergangswahrscheinlichkeiten unterscheiden. Die <b>beobachteten relativen Häufigkeiten können wir dann als Näherung für die Übergangswahrscheinlichkeiten verwenden</b>.
<br><br>
Dafür müssen allerdings geeignete Trainingsdatensätze ausgewählt werden, da die relativen Häufigkeiten stark vom vorliegenden Text abhängen. In unserem Beispiel würde die Übergangswahrscheinlichkeit von "mag" zu "Mathe" zu null genähert werden, obwohl diese Wortsequenz im Sprachgebrauch sicherlich vorkommen kann. Um die Näherung der Übergangswahrscheinlichkeiten durch die relativen Häufigkeiten zu verbessern, müssen umfangreichere Trainingsdatensätze verwendet werden, die den Sprachgebrauch beim Chatten möglichst gut repräsentieren. Wir verwenden daher einen größeren Datensatz von dem wir annehmen, dass er Alltagstexte ausreichend gut repräsentiert.
    </ol>    
</div><!-------In unserem Modell fassen wir das Auswählen eines Wortes nach einem bereits geschriebenen Wort als Zufallsexperiment auf. Bei vorgegebenem ersten Wort "ich" können wir uns dieses Experiment als das Ziehen aus einer Urne vorstellen. Die Urne enthält alle Zweier-Wortsequenzen aller Texte, die beim Chatten verfasst wurden und die mit dem Wort "ich" beginnen. In diesem Zufallsexperiment gibt es Wahrscheinlichkeiten für die Ausgänge des Zufallsexperimentes. Diese Wahrscheinlichkeit können wir nicht herausfinden. Wir hoffen, dass die relativen Häufigkeiten mit der Wahrscheinlichkeit mit der ein Wort nach einem bereits getippten Wort ausgewählt wird gut übereinstimmen. Wir treffen somit die folgende <b>Modellannahme</b>: Wir nehmen an, dass der Trainingsdatensatz die Gesamtheit aller Texte, die beim Chatten verfasst werden, möglichst gut repräsentiert. --->
